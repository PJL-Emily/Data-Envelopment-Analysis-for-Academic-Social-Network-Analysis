```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import time
import random
```

```python
volumes = range(256, 296)
issues = [1, 2, 3]
basic_url = 'https://www.sciencedirect.com/journal/european-journal-of-operational-research/vol/'
```

```python
# df = pd.DataFrame(columns=['title', 'author', 'affiliation', 'author idx'])
# df_url = pd.DataFrame(columns=['volume', 'issue', 'url'])
df = pd.read_csv('EJOR_vol_256-295.csv', index_col=False)
df_url = pd.read_csv('EJOR_url.csv', index_col=False)

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
for v in range(285, 296, 3): # volumes:
    for iss in issues:
        driver.get(basic_url + str(v) + '/issue/' + str(iss))
        articles = driver.find_elements(By.CLASS_NAME, 'js-article-title')
        # print(len(articles))
        for i in range(1, len(articles)):
            time.sleep(random.uniform(0,2))
            ## Enter the page of each article ##
            articles = driver.find_elements(By.CLASS_NAME, 'js-article-title')
            articles[i].click()
            driver.find_element(By.ID, 'show-more-btn').click() # show affiliation
            df_url = df_url.append({'volume': v, 'issue': iss, 'url': driver.current_url}, ignore_index=True)
            try:
                ## Get content ##
                title = driver.find_element(By.CSS_SELECTOR, "span[class='title-text']").text
                authors = driver.find_elements(By.CSS_SELECTOR, "span[class='content']")
                affiliates = driver.find_elements(By.CSS_SELECTOR, "dl[class='affiliation']")
                # print(title, authors, affiliates)
                if len(affiliates) > 1:
                    code2aff = {}
                    for aff in affiliates:
                        code = aff.find_element(By.CSS_SELECTOR, 'sup').text
                        name = aff.find_element(By.CSS_SELECTOR, 'dd').text
                        code2aff[code] = name
                    # print(code2aff)
                cnt = 0
                for author in authors:
                    cnt += 1
                    fName = author.find_element(By.CSS_SELECTOR, "span[class='text given-name']").text
                    lName = author.find_element(By.CSS_SELECTOR, "span[class='text surname']").text
                    if len(affiliates) > 1:
                        affcod = author.find_element(By.CSS_SELECTOR, 'sup').text # only counts the first affiliation
                        affil = code2aff[affcod]
                        # print(title, fName, lName, affil)
                        df = df.append({'title': title, 'author': fName + ' ' + lName, 'affiliation': affil, 'author idx': cnt}, ignore_index=True)
                    else:
                        affil = affiliates[0].find_element(By.CSS_SELECTOR, 'dd').text
                        # print(title, fName, lName, affil)
                        df = df.append({'title': title, 'author': fName + ' ' + lName, 'affiliation': affil, 'author idx': cnt}, ignore_index=True)
            except:
                print('Error:', title)
            driver.back() # Return to the page of all artucles
    df.to_csv('EJOR_vol_256-295.csv', index=False)
    df_url.to_csv('EJOR_url.csv', index=False)
```
